```python
class Person:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_info(self):
        print(f"Name: {self.name}, Phone: {self.phone}")


class Patient(Person):
    def __init__(self, name, age, phone, disease):
        super().__init__(name, phone)
        self.age = age
        self.disease = disease
        self.reports = []

    def add_report(self, report):
        self.reports.append(report)


class Doctor(Person):
    def __init__(self, name, phone, speciality):
        super().__init__(name, phone)
        self.speciality = speciality

    def prescribe(self):
        print(f"Dr. {self.name} is writing a prescription.")


class Appointment:
    def __init__(self, patient, doctor, date):
        self.patient = patient
        self.doctor = doctor
        self.date = date

    def details(self):
        print(f"Appointment: {self.patient.name} with Dr. {self.doctor.name} on {self.date}")


class Report:
    def __init__(self, patient, report_name):
        self.patient = patient
        self.report_name = report_name

    def show_report(self):
        print(f"Report: {self.report_name} for {self.patient.name}")


# ----- Taking Patient Info -----
p_name = input("Enter patient name: ")
p_age = int(input("Enter patient age: "))
p_phone = input("Enter patient phone: ")
p_disease = input("Enter patient disease: ")

patient = Patient(p_name, p_age, p_phone, p_disease)


# ----- Taking Doctor Info -----
d_name = input("Enter doctor name: ")
d_phone = input("Enter doctor phone: ")
d_speciality = input("Enter doctor speciality: ")

doctor = Doctor(d_name, d_phone, d_speciality)


# ----- Appointment -----
appt_date = input("Enter appointment date (YYYY-MM-DD): ")

appointment = Appointment(patient, doctor, appt_date)
appointment.details()


# ----- Add Report -----
report_name = input("Enter report name: ")
report = Report(patient, report_name)

patient.add_report(report)
report.show_report()
```
